```python
#Exp 2 - Amicable Numbers
def amicable_number(num1,num2):
    sum1 =0
    sum2=0
    for i in range(1,num1):
        if num1%i ==0:
            sum1+= i

    for i in range(1,num2):
        if num2%i==0:
            sum2+=i

    if sum1==num2 and sum2==num1:
        print("these are amicable")
    else:
        print("these are not amicable")

amicable_number(1184,1210)
```

```python
#Exp3 - Twin Prime
def Isprime(num):
    if num<=1:
        return False
    for i in range(2,num):
        if num %i ==0:
            return False
        else:
            return True


def twin_prime(limit):
    for i in range(2,limit):
        if Isprime(i) and Isprime(i+2):
            print(f"({i},{i+2})")

twin_prime(1000)
```

```python
#Exp 7
#Box plot - wisker plot,minimum,median,first quartile,third quartile,outliers
qws=np.random.normal(10,30,400) #where 10 is mean 30 is standard deviation and 400 is random variables 
plt.boxplot(qws)
plt.show()

#Bee swarm Plot - individual data point without overlapping
import seaborn as sns
sns.swarmplot(qws)
plt.show()

# Violin Plot - more info than boxplot, full distribution of data
sns.violinplot(qws)
plt.show()
```
