```python
# using pandas
import pandas as pd
file = pd.read_csv("cars.csv")
file
```

```python
#using open
with open("cars.csv") as file:
    for line in file.readlines():
        print(line)
```

```python
#Setting up chromadb
import chromadb
client = chromadb.Client()
#Creating verctorDB database
collection = client.create_collection(name="cars")
```

```python
# creating data in a the required fields to apend in db
ids =[]
metadata =[]
document =[]

for i,row in file.iterrows():
    doc = f"{row['Car']} ({row['Year']}) - Color: {row['Color']} , Price:{row['Price ($)']}"
    document.append(doc)
    metadata.append({'Car':row['Car'],
                     'Year':row['Year'],
                     'Color':row['Color'],
                     'Price':row['Price ($)']
                     })
    ids.append(str(i))
```

```python
# feeding the data to vector db database
collection.add(
    ids=ids,
    documents=document,
    metadatas=metadata
)
```

```python
# querying from db
query = "Which car is electric?"
result = collection.query(
             query_texts=[query],
            n_results=2
            )

print("query", query)
print("result", result['documents'])
```
