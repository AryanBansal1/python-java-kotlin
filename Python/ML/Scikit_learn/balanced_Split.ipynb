```python
from sklearn.datasets import load_iris
data = load_iris()
x,y = load_iris(return_X_y=True)
```

```python
#Unbalanced distribution of example of each cateogry of label into test , training
from sklearn.model_selection import train_test_split
trainX,testX,trainY,testY = train_test_split(x,y,test_size=0.2)  #keeping testing size 20%
```

```python
# to show that it has unbalanced distribution of Training data
import numpy as np, matplotlib.pyplot as plt

counts = np.bincount(trainY)
positions = np.arange(3)
plt.bar(positions,counts)
plt.xticks(positions,data.target_names)
```

```python
# to show the unbalanced distribution of testing data
counts = np.bincount(testY)
positions = np.arange(3)
plt.bar(positions,counts)
plt.xticks(positions,data.target_names)
```

```python
# lets check the previous score
from sklearn.neighbors import KNeighborsClassifier
mod = KNeighborsClassifier()
mod.fit(trainX,trainY)
print(mod.score(testX,testY))
```

```python
# To solve this problem we have something that will distrubte the data equally
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1,test_size=0.2)

for train_idx , test_idx in split.split(x,y):
    trainX, testX = x[train_idx], x[test_idx] 
    trainY,testY = y[train_idx],y[test_idx]
```

```python
counts = np.bincount(testY)
positions = np.arange(3)
plt.bar(positions,counts)
plt.xticks(positions, data.target_names)
```

```python
mod = KNeighborsClassifier()
mod.fit(trainX,trainY)
print(mod.score(testX,testY))
```
